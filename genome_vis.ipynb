### Add the track
User can freely place tracks within the sector radius range (0 - 100). 
Adding tracks to a Circos plot allows you to display multiple layers of information within each sector. This is particularly useful when you want to visualize complex and multi-dimensional data.

Each track can represent different data or different aspects of the same data. For example, in a genomic Circos plot, different tracks could represent gene expression levels, mutation sites, and structural variations, all in the same plot.

By using tracks, you can compare and contrast these different data layers directly, making it easier to identify patterns and correlations that might not be apparent when looking at the data separately. This makes Circos plots a powerful tool for data exploration and analysis.



```python
from pycirclize import Circos

# Initialize circos sectors
sectors = {"A": 10, "B": 15, "C": 12, "D": 20, "E": 15}
circos = Circos(sectors, space=5)

for sector in circos.sectors:
    # Plot sector axis & name text
    sector.axis(fc="none", ls="dotted", lw=2, ec="black", alpha=0.8)
    sector.text(f"Sector: {sector.name}={sector.size}", size=15)
    # Set Track01 (Radius: 85 - 100)
    track1 = sector.add_track((85, 100))
    track1.axis(fc="blue", alpha=0.5)
    track1.text(track1.name)
    # Set Track02 (Radius: 75 - 84)
    track2 = sector.add_track((75, 84))
    track2.axis(fc="brown", alpha=1)
    track2.text(track2.name)
    # Set Track03 (Radius: 60 - 74)
    track3 = sector.add_track((60, 74))
    track3.axis(fc="orange", alpha=1)
    track3.text(track3.name)
    # Set Track04 (Radius: 40 - 59)
    track4 = sector.add_track((40, 59))
    track4.axis(fc="purple", alpha=1)
    track4.text(track4.name)
    # Set Track05 (Radius: 0 - 35)
    track5 = sector.add_track((0, 35))
    track5.axis(fc="lime", alpha=0.7)
    track5.text(track5.name)

fig = circos.plotfig()
```

The for loop iterates over each sector in the Circos object:

sector.axis(fc="none", ls="dotted", lw=2, ec="black", alpha=0.8): This line plots the axis of the current sector with specified properties.
sector.text(f"Sector: {sector.name}={sector.size}", size=15): This line adds a text label to the sector, displaying its name and size.
track1 = sector.add_track((85, 100)): This line adds a track to the current sector with a radius range from 85 to 100. Similar steps are repeated for track2, track3, track4, and track5 with different radius ranges and face colors.
fig = circos.plotfig(): This line generates the final Circos plot and assigns it to the fig variable.

The benefit of tracks in a Circos plot is that they allow you to display multiple layers of information within each sector. Each track can represent different data or different aspects of the same data. This makes Circos plots a powerful tool for visualizing complex and multi-dimensional data.



```python
from pycirclize import Circos
import numpy as np
np.random.seed(0)

sectors = {"A": 10, "B": 15, "C": 12, "D": 20, "E": 15}
circos = Circos(sectors, space=5)
for sector in circos.sectors:
    # Plot sector name
    sector.text(f"Sector: {sector.name}", r=110, size=15)
    # Create x positions & randomized y values for data plotting
    x = np.arange(sector.start, sector.end) + 0.5
    y = np.random.randint(0, 100, len(x))
    # Plot line
    line_track = sector.add_track((75, 100), r_pad_ratio=0.1)
    line_track.axis()
    line_track.xticks_by_interval(1)
    line_track.line(x, y)
    # Plot points
    points_track = sector.add_track((45, 70), r_pad_ratio=0.1)
    points_track.axis()
    points_track.scatter(x, y)
    # Plot bar
    bar_track = sector.add_track((15, 40), r_pad_ratio=0.1)
    bar_track.axis()
    bar_track.bar(x, y)

fig = circos.plotfig()
```

```python
from pycirclize import Circos
from pycirclize.parser import Gff
from pycirclize.utils import load_prokaryote_example_file

# Load GFF file
gff_file = load_prokaryote_example_file("enterobacteria_phage.gff")
gff = Gff(gff_file)

circos = Circos(sectors={gff.name: gff.range_size})
circos.text("Enterobacteria phage\n(NC_000902)", size=15)

sector = circos.sectors[0]
cds_track = sector.add_track((90, 100))
cds_track.axis(fc="#EEEEEE", ec="none")

# Plot forward & reverse CDS
f_cds_feats = gff.extract_features("CDS", target_strand=1)
cds_track.genomic_features(f_cds_feats, plotstyle="arrow", r_lim=(95, 100), fc="salmon")
r_cds_feats = gff.extract_features("CDS", target_strand=-1)
cds_track.genomic_features(r_cds_feats, plotstyle="arrow", r_lim=(90, 95), fc="skyblue")

# Extract CDS product labels
pos_list, labels = [], []
for feat in gff.extract_features("CDS"):
    start, end = int(str(feat.location.end)), int(str(feat.location.start))
    pos = (start + end) / 2
    label = feat.qualifiers.get("product", [""])[0]
    if label == "" or label.startswith("hypothetical"):
        continue
    if len(label) > 20:
        label = label[:20] + "..."
    pos_list.append(pos)
    labels.append(label)

# Plot CDS product labels on outer position
cds_track.xticks(
    pos_list,
    labels,
    label_orientation="vertical",
    show_bottom_line=True,
    label_size=6,
    line_kws=dict(ec="grey"),
)
# Plot xticks & intervals on inner position
cds_track.xticks_by_interval(
    interval=5000,
    outer=False,
    show_bottom_line=True,
    label_formatter=lambda v: f"{v/ 1000:.1f} Kb",
    label_orientation="vertical",
    line_kws=dict(ec="grey"),
)

fig = circos.plotfig()
```

# Ideogram
An ideogram is a graphical representation or visual depiction of the karyotype, which is the complete set of chromosomes in an organism's genome. Ideograms are commonly used in genetics and cytogenetics to illustrate the characteristic features of chromosomes, such as their size, shape, banding patterns, and genetic content.

In an ideogram, each chromosome is typically represented as a straight or curved line, with specific regions and features indicated along its length. These features may include:

1. **Centromeres**: The central region of the chromosome where sister chromatids are joined and spindle fibers attach during cell division.

2. **Telomeres**: The protective caps at the ends of chromosomes that help maintain chromosome stability and integrity.

3. **Band Patterns**: The distinctive patterns of light and dark bands that are visible when chromosomes are stained and viewed under a microscope. These bands correspond to different regions of DNA with varying densities of genes and other DNA sequences.

4. **Genetic Markers**: Specific genes, genetic loci, or DNA sequences that are of interest for genetic analysis or research purposes.

5. **Structural Aberrations**: Anomalies or abnormalities in chromosome structure, such as deletions, duplications, inversions, or translocations.

Ideograms provide a visual reference for geneticists and researchers to study and analyze the chromosomal composition of an organism's genome. They are particularly useful for identifying chromosomal abnormalities associated with genetic disorders, mapping genes and genetic traits, and studying the evolutionary relationships between species.



```python
from pycirclize import Circos
from pycirclize.utils import load_eukaryote_example_dataset

# Load hg38 dataset (https://github.com/moshi4/pycirclize-data/tree/main/eukaryote/hg38)
chr_bed_file, cytoband_file, _ = load_eukaryote_example_dataset("hg38")

# Initialize Circos from BED chromosomes
circos = Circos.initialize_from_bed(chr_bed_file, space=6)
circos.text("Homo sapiens (hg38)", size=20)

# Add cytoband tracks from cytoband file
circos.add_cytoband_tracks((93, 100), cytoband_file)

# Plot chromosome name
for sector in circos.sectors:
    sector.text(sector.name, size=15)

fig = circos.plotfig()
```

1. Import the necessary modules: Circos from pyCirclize and load_eukaryote_example_dataset from pyCirclize.utils.

2. Load the hg38 dataset using the load_eukaryote_example_dataset function. This function returns three values: the path to the BED file containing the chromosome data, the path to the cytoband file containing the cytoband data, and a third value that is not used in this script.

3. Initialize a Circos object from the BED file using the initialize_from_bed method. The space parameter is set to 6, which determines the space between the sectors in the plot.

4. Add a text label to the plot using the text method. The label is "Homo sapiens (hg38)" and the font size is 20.

5. Add cytoband tracks to the plot using the add_cytoband_tracks method. The first parameter is a tuple specifying the inner and outer radii of the tracks, and the second parameter is the path to the cytoband file.

6. Loop over the sectors in the Circos object and add a text label to each sector using the text method. The label is the name of the sector and the font size is 15.

7. Finally, plot the Circos object using the plotfig method. The resulting figure is stored in the fig variable.

The resulting plot is a circular representation of the human genome, with each chromosome represented as a sector and the cytobands represented as tracks within each sector. The name of each chromosome is labeled on the plot.



# Segmental Dups Link

Segmental duplications (also known as segmental duplications or segmental duplications) are regions of DNA that are duplicated within a genome and are larger than typical duplication events, often spanning tens to hundreds of kilobases in length. These duplications can be located on the same chromosome (intra-chromosomal) or on different chromosomes (inter-chromosomal).

The Segmental Dups Link likely refers to a link or connection between these segmental duplications, indicating the presence of duplicated genomic segments and potentially providing information about their locations, lengths, and relationships. This link may be visualized in various genomic analysis tools or databases to illustrate the presence and arrangement of segmental duplications within a genome.



```python
# Import required modules
from pycirclize import Circos
from pycirclize.utils import ColorCycler, load_eukaryote_example_dataset

# Load example dataset for human genome assembly hg38
chr_bed_file, cytoband_file, chr_links = load_eukaryote_example_dataset("hg38")

# Initialize Circos plot using chromosome BED file data
circos = Circos.initialize_from_bed(chr_bed_file, space=3)
# Add text annotation indicating the genome being plotted
circos.text("Homo sapiens\n(hg38)", deg=315, r=150, size=12)

# Add cytoband tracks to the Circos plot using cytoband file data
circos.add_cytoband_tracks((95, 100), cytoband_file)

# Set color map to HSV (Hue, Saturation, Value) for defining chromosome colors
ColorCycler.set_cmap("hsv")

# Define chromosome colors
chr_names = [s.name for s in circos.sectors]
colors = ColorCycler.get_color_list(len(chr_names))
chr_name2color = {name: color for name, color in zip(chr_names, colors)}

# Plot chromosome names and xticks
for sector in circos.sectors:
    # Add chromosome names
    sector.text(sector.name, r=120, size=10, color=chr_name2color[sector.name])
    # Add xticks for cytoband track
    sector.get_track("cytoband").xticks_by_interval(
        40000000,
        label_size=8,
        label_orientation="vertical",
        label_formatter=lambda v: f"{v / 1000000:.0f} Mb",
    )

# Plot chromosome links
for link in chr_links:
    region1 = (link.query_chr, link.query_start, link.query_end)
    region2 = (link.ref_chr, link.ref_start, link.ref_end)
    color = chr_name2color[link.query_chr]
    # Add links between chromosomes to the Circos plot
    if link.query_chr in ("chr1", "chr8", "chr16") and link.query_chr != link.ref_chr:
        circos.link(region1, region2, color="PaleTurquoise",alpha=1)

# Generate Circos plot
fig = circos.plotfig()
```
